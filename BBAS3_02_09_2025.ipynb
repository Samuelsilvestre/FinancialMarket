```python
import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python

ticker = "BBAS3.SA"

data_inicio = "2025-01-01"
data_fim = "2025-09-02"
bb = yf.download(ticker, start=data_inicio, end=data_fim)
bb = bb['Close']
display(bb.head(20))
```

```python
bb.columns = ['Fechamento']
```

Preço de fechamento ao longo do ano

```python
sns.set_style('darkgrid')
plt.figure(figsize=(18,8))
sns.lineplot(bb)
plt.title('BBAS3 ao longo do ano', fontsize=20)
plt.xlabel('Data')
plt.ylabel('Preço')
plt.show();
```

# Média movel

```python
bb['Media Movel'] = bb['Fechamento'].rolling(20).mean()
```

```python
sns.set_style('darkgrid')
plt.figure(figsize=(18,8))
sns.lineplot(bb)
plt.title('BBAS3 média movel', fontsize=20)
plt.xlabel('Data')
plt.ylabel('Preço')
plt.show();
```

# Média movel trimestral

```python
bb['Media movel Trimestral'] = bb['Fechamento'].rolling(90).mean()
```

```python
sns.set_style('darkgrid')
plt.figure(figsize=(18,8))
sns.lineplot(bb[['Media movel Trimestral', 'Fechamento']])
plt.title('BBAS3 média movel trimestral' , fontsize=20)
plt.xlabel('Data')
plt.ylabel('Preço')
plt.show();
```

### Conclusão sobre o Cenário do Banco do Brasil

O cenário para o Banco do Brasil (BBAS3) é **incerto**, impactado por fatores externos à sua gestão.

Os principais motivos para essa incerteza são:

* **Inadimplência no Agronegócio**: Aumento de dívidas no setor, que pode afetar a qualidade de crédito do banco.
* **Possíveis Sanções Internacionais**: O risco de sanções do governo americano contra o governo brasileiro e suas estatais gera instabilidade e pode afetar a confiança dos investidores.

Apesar de bons resultados, esses fatores externos podem pesar sobre o desempenho futuro das ações do banco.
